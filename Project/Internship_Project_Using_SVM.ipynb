# Image Processing

```python
labels={"Abhay":0 ,"Atul":1, "Deepak":2, "Ketan":3, "Prateek":4, "Pritam":5, "Sanjeev":6, "Swapan":7 }
img_data=[]
labelList=[]
```

```python
for folder in directory:
    label=os.path.basename(folder)

    for images in folder.glob("*.jpg"):
        img=load_img(images,target_size=(128,128))
        im_array=img_to_array(img)
        img_data.append(im_array)
        labelList.append(labels[label])
```

```python
img_data=np.array(img_data)
img_data
```

# Data Augmetation

```python
datagen=ImageDataGenerator()
```

# Creating Model

```python
from sklearn import svm
```

```python
model=svm.SVC(kernel='poly',degree=3,C=1,gamma=0.001,probability=True)
```

```python
#history= model.fit_generator(datagen.flow(trainX,trainY,batch_size=10),epochs=20,validation_data=(testX,testY))
```

```python
epochs=25
for val in range(epochs):
    print("epoch: ", val)
    b=0
    for batchX,batchY  in datagen.flow(trainX,trainY):
        model.fit(batchX.reshape(len(batchX),-1),batchY)
        b+=1
        if b>=len(trainX):     
                  break
```

```python
prediction=model.predict(testX.reshape(46,49152))
(prediction,np.transpose(testY))
```

```python
from sklearn.metrics import accuracy_score
ac=accuracy_score(testY,prediction)*100
print("Accuracy : ",ac,"%")
```
